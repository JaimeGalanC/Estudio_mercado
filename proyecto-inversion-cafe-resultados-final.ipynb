### Nombres de cadenas comerciales

Tratemos de homologar los nombres de los establecimientos en la columna `'object_name'` para que los establecimientos que pertenezcan a una cadena muestren el mismo nombre.

```python
"""Definamos una función para extraer el nombre de un establecimiento que pueda pertenecer a una cadena"""
def extraer_pre_name_chain(object_name):
    """
    object_name=nombre del establecimiento
    """
    nombre=object_name
    
    # eliminemos los digitos y los numerales al inicio y final del texto
    patron = re.compile(r'#\d+|#\s\d+|#\w+\d+\-\d+')
    texto_limpio = patron.sub('', nombre)
    
    # vamos a reemplazar 'CO.' por 'COMPANY'
    patron = re.compile(r'COMPANY')
    texto_limpio_2 = patron.sub('CO.', texto_limpio)
    
    # vamos a reemplazar 'INCORPORATION' por 'INC.'
    patron = re.compile(r'INCORPORATION')
    texto_limpio_3 = patron.sub('INC.', texto_limpio_2).strip()
    
    return texto_limpio_3
```

```python
# aplicamos la funcion para extraer el 'pre_name_chain' de los establecimientos
restaurant_data['pre_name_chain']=restaurant_data['object_name'].apply(extraer_pre_name_chain)
```

### Separar nombres de calles de la dirección  
Para hacer esta separación usaremos el método `'re.compile'` que nos ayuda a separar un texto. De la biblioteca `'usaddress'` vamos a utilizar los datos de los tipos de direcciones ('Norte', 'Sur, etc) y abreviaciones de los tipos de calles ('St','Street', etc), `'DIRECTIONS'` y `'STREET_NAMES'`, respectivamente.

```python
# creamos una lista con los tipos de direcciones usando 'DIRECTIONS'
directions_list = [valor.upper() for valor in DIRECTIONS]
```

```python
# creamos una lista con los tipoa de calles usando 'STREET_NAMES'
street_names_list =[valor.upper() for valor in STREET_NAMES]

"""Añadimos un valor a remover o añadir en la lista de 'street_names_list', esto nos va permitir
que haya menos calles con valores vacios o nulos, o que el nombre aparezca más limpio"""
street_names_list.append('STE')
street_names_list.append('PK')
street_names_list.remove('HILL')
street_names_list.remove('WAY')
street_names_list.remove('GROVE')
street_names_list.remove('PARK')
street_names_list.remove('SPRING')
street_names_list.remove('GATEWAY')
street_names_list.remove('VALLEY')
street_names_list.remove('UNION')
street_names_list.remove('CENTER')
street_names_list.remove('FOREST')
street_names_list.remove('WALL')
street_names_list.remove('MISSION')
street_names_list.remove('FORD')
street_names_list.remove('AVENUE')
street_names_list.remove('PLAZA')
street_names_list.remove('VILLAGE')
street_names_list.remove('TERRACE')
```

```python
"""Creamos la función para extraer el nombre de la calle a partir de una dirección en USA, usando re.compile"""
def extraer_pre_calle(address):
    """
    address=la direccion en USA
    """
    domicilio = address
    
    # direcciones
    direcciones=['E','W','S','N','SE','SW','NE','NW']
    
    # eliminemos los digitos y los numerales al inicio y final del texto
    nuevo_texto = re.sub(r'^\d+|\s\d+\s|\s#.*|\s\w\d+','', domicilio)
    
    # quitamos abreviaturas de direcciones
    pattern1 = '|'.join(map(re.escape, direcciones))+" "
    p=re.compile(r'\b(?:{})\b'.format(pattern1))
    sin_direcciones=p.sub('',nuevo_texto).strip()
    
    # quitamos abreviaturas de calles
    pattern2 = '|'.join(map(re.escape, street_names_list))
    
    p=re.compile(r'\b(?:{})\b'.format(pattern2))
    sin_tipo_calle=p.sub('',sin_direcciones)
    
    # quitamos algun digito que haya quedado
    resultado = re.sub(r'\s\d+$', '', sin_tipo_calle).strip()
    
    return resultado
```

```python
# aplicamos la funcion para extraer el nombre de la pre_calle
restaurant_data['pre_calle']=restaurant_data['address'].apply(extraer_pre_calle)
```

```python
# mostramos parte de los datos
restaurant_data.sample(5)
```

```python
# contamos las veces que se repite una calle y mostramos percentiles para este conteo
data=restaurant_data['pre_calle'].value_counts()
np.percentile(data,[50,90,95,97,99])
```

```python
# creamos una lista con las calles que se repiten 22 ó mas veces
lista_calles=[data.index[i] for i in range(len(data)) if data[i]>=22]
```

```python
"""Creamos una columna con el nombre de las calles definitivas"""
def extraer_calle(pre_calle):
    """
    pre_calle= la columna que contiene los valores a usar para crear la columna de calle definitiva
    """
    
    # comparamos los valores en la lista 'lista_calles' con los valores en la columna 'pre_calle'
    for elem in lista_calles:
        if elem in pre_calle:
            return elem
        else:
            return pre_calle
```

```python
# aplicamos la funcion para extraer el nombre de la calle
restaurant_data['calle']=restaurant_data['pre_calle'].apply(extraer_calle)
```

```python
# eliminamos la columna 'pre_calle'
restaurant_data.drop('pre_calle',inplace=True,axis=1)
```

### Proporciones de los distintos tipos de establecimientos 
Para estudiar las proporciones de los distintos tipos de establecimientos vamos a agrupar el set de datos por la columna `'object_type'`.

```python
# obtenemos las proporciones para cada tipo de establecimiento
proporcion_object_type=restaurant_data['object_type'].value_counts(normalize=True)\
                            .reset_index()
proporcion_object_type.columns=['category','proportion']
proporcion_object_type
```

```python
# Trazamos un gráfico de las proporciones de los tipos de establecimiento
f,ax=plt.subplots(figsize=(12,6))
ax=sns.barplot(data=proporcion_object_type,x='category',y='proportion',\
            order=proporcion_object_type.sort_values('proportion',ascending = False).category)

# add text
ax.text(3,0.7,"Hay una gran cantidad de restaurantes establecidos en los L.A.",\
            fontsize = 14,          # Tamaño
            fontstyle = "normal",  # Estilo
            color = "black",ha="center",va="center")
ax.text(2.8,0.6,"Un establecimiento de café está en el tercer lugar con presencia en los L.A.",\
            fontsize = 14,          # Tamaño
            fontstyle = "normal",  # Estilo
            color = "black",ha="center",va="center")
ax.text(0.5,-0.25,"Fuente: datos abiertos sobre restaurantes en los L.A.",\
            fontsize = 10,          # Tamaño
            fontstyle = "normal",  # Estilo
            color = "black",ha="center",va="center")
# For each bar: Place a label
for rect in ax.patches:
    # Get X and Y placement of label from rect.
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2

    # Number of points between bar and label
    space = 3
    # Vertical alignment 'va'

    # Use Y value as label and format number with zero decimal place
    label = "{:.1%}".format(y_value)

    # Create annotation
    ax.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(0, space),          # Vertically shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        ha='center',                # Horizontally center label
        va='bottom')     
plt.ylim(0,proporcion_object_type['proportion'].max()*1.1)
plt.title('Proporciones por establecimiento')
plt.ylabel('Proporción')
plt.xlabel('Establecimiento')
plt.xticks(rotation=45,ha='right')
plt.show()
```

### ¿Cuántas cafeterías hay en las diez calles con más establecimientos?  
Vamos a hacer uso del dataframe `'best_calles'` que contiene los nombres de las diez calles con más establecimientos.

```python
# contamos la cantidad de cafeterias en las diez calles con mas establecimientos
count_cafe=restaurant_data[(restaurant_data['calle'].isin(best_calles.index)) & (restaurant_data['object_type']=='CAFE')]['calle']\
                                            .value_counts()
count_cafe.name='count_cafe'
count_cafe
```

```python
# unimos 'best_calles' y 'count_cafe' para ver la proporcion de cafeterias
proportion_cafe=best_calles.merge(count_cafe,left_on=best_calles.index,right_on=count_cafe.index)
proportion_cafe.columns=['calle','count_type','count_cafe']

# obtenemos la proporcion de cafeterias respecto a la cantidad de establecimientos
proportion_cafe['proportion_cafe']=proportion_cafe['count_cafe']/proportion_cafe['count_type']

# aplicamos formato de porcentaje
proportion_cafe['proportion_cafe']=proportion_cafe['proportion_cafe'].map('{:.1%}'.format)
proportion_cafe
```

**En la tabla anterior, podemos observar que la proporción de cafeterías existentes en las principales calles con más establecimientos es muy baja. Así que eso puede representarnos una oportunidad de poder establecer una cafetería en las calles con menor proporción de cafeterías y que sean muy transitadas.**
